```python
# billing system

def calculate_bill(prices):  # function
    total_bill = sum(prices)
    average_price = total_bill / len(prices) if prices else 0
    # Get top 3 expensive items
    top_three = tuple(sorted(prices, reverse=True)[:3])

    return total_bill, average_price, top_three

sample_prices = [35.50, 15.75, 42.99, 18.25, 36.40, 19.99, 87.00]
total, average, top_items = calculate_bill(sample_prices)

print("Store Billing System")
print(f"Sample prices: {sample_prices}")
print(f"Total bill: ${total:.2f}")
print(f"Average item price: ${average:.2f}")
print(f"Three most expensive items: {top_items}")
print()
```

```python
def process_sentence(sentence):
    # 1. Count characters excluding spaces
    chars_without_spaces = len(sentence.replace(" ", ""))
    # 2. Convert to uppercase and lowercase
    upper_case = sentence.upper()
    lower_case = sentence.lower()
    # 3. Find first and last word
    words = sentence.split()
    first_word = words[0] if words else ""
    last_word = words[-1] if words else ""
    # 4. Replace "Python" with "AI"
    modified_sentence = sentence.replace("Python", "AI")
    # 5. Display results
    print("String Operations")
    print(f"Original sentence: {sentence}")
    print(f"1. Total characters (excluding spaces): {chars_without_spaces}")
    print(f"2. Uppercase: {upper_case}")
    print(f"   Lowercase: {lower_case}")
    print(f"3. First word: '{first_word}', Last word: '{last_word}'")
    print(f"4. Modified sentence: {modified_sentence}")
    print(f"5. Modified sentence length: {len(modified_sentence)}")

sentence = input("Enter a sentence: ")
process_sentence(sentence)
print()
```

```python
def find_top_student(students):
    if not students:
        return None, None
    top_student = max(students.items(), key=lambda x: x[1])
    return top_student

print("Student Management System")
student_marks = {"Ali": 85, "Sara": 92, "Ahmed": 78, "Fatima": 95, "Khalid": 88}
top_name, top_marks = find_top_student(student_marks)

print(f"Student marks: {student_marks}")
print(f"Top performing student: {top_name} with {top_marks} marks")
print()
```

```python
def calculate_salary(basic_salary):
    # Apply 10% performance bonus
    salary_with_bonus = basic_salary * 1.10

    # Apply 5% tax deduction
    final_salary = salary_with_bonus * 0.95

    return final_salary

print("Salary Calculation")
salaries = [50000, 75000, 30000]

for basic in salaries:
    final = calculate_salary(basic)
    print(f"Basic salary: ${basic:,.2f}")
    print(f"After 10% bonus: ${basic * 1.10:,.2f}")
    print(f"Final salary (after 5% tax): ${final:,.2f}")
    print(f"---")
```

```python
def string_slicing(text):

    print("String Slicing Operations")
    print(f"Original string: '{text}'")

    # 1. First 5 characters
    print(f"1. First 5 characters: '{text[:5]}'")

    # 2. Last 4 characters
    print(f"2. Last 4 characters: '{text[-4:]}'")

    # 3. String in reverse order
    print(f"3. Reverse order: '{text[::-1]}'")

    # 4. Every second character
    print(f"4. Every second character: '{text[::2]}'")

    # 5. Substring from index 3 to 10
    print(f"5. Substring from index 3 to 10: '{text[3:11]}'")

sample_text = "Applied Programming"
string_slicing(sample_text)
```
